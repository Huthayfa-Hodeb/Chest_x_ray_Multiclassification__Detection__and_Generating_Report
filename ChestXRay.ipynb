# Reading CSV files into pandas DataFrames

```python
bb_data = pd.read_csv("Data/BBox_List_2017.csv")
data = pd.read_csv("Data/Data_Entry_2017.csv")
```

# Extract B_Box images

```python
for i in range(bb_data.shape[0]):
    path = f"D:\\pr\\images_003\\images\\{bb_data['Image Index'][i]}"
    img = cv2.imread(path)
    img_arr = np.asarray(img)
    if img_arr.sum() == None:
        continue
    else:
        cv2.imwrite(f"Data/bbox_img/{bb_data['Image Index'][i]}", img)
```

# Joining B_Box Data with Data_Entry

```python
bb_data = bb_data.set_index('Image Index')
data = data.set_index('Image Index')
```

```python
bb_new = bb_data.join(data, how = "inner")
```

```python
bb_new = bb_new.reset_index()
```

# Create a CSV file for annotating images when uploading them to Roboflow.

```python
bb_new.head()
```

```python
robo_format = bb_new.iloc[:,[0, 1, 2, 3]]
```

```python
robo_format.head()
```

```python
robo_format["xmax"] = robo_format["Bbox [x"] + bb_new["w"]
robo_format["ymax"] = robo_format["y"] + bb_new["h]"]
robo_format["width"] = 1024
robo_format["height"] = 1024
```

```python
robo_format = robo_format.rename(columns = {"Image Index":"filename", "Finding Label":"class", "Bbox [x":"xmin", "y":"ymin"})
```

```python
robo_format = robo_format.iloc[:, [0,6,7,1,2,3,4,5]]
```
